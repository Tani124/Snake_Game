# Detailed Version

#### Project Roadmap

Step 2: Create Snake Block - Snake variables

Step 3: Snake Movement - Based on dx, dy

Step 4: Continuos Movement - Movement will keep happening - Snake will not stay still

Step 6: Food Generation - Function for food - Use random for spawning

Step 7: Eating Food - Collision with food

Step 8: Collision Detection - Walls & Self

Step 9: Game Over Screen - In case of collision - Game Over

```python
import pygame 
import random

pygame.init()

WIDTH = 600
HEIGHT = 600 
screen = pygame.display.set_mode([WIDTH, HEIGHT])
pygame.display.set_caption("Snake game")
clock = pygame.time.Clock()

block_size = 20
snake_x = WIDTH // 2
snake_y = HEIGHT // 2

dx = 0 
dy = 0 

snake_body = []
snake_length = 1

def spawn_food():
    fx = round(random.randrange(0, WIDTH - block_size) / 20.0) * 20.0
    fy = round(random.randrange(0, HEIGHT - block_size) / 20.0) * 20.0
    return fx, fy

food_x, food_y = spawn_food()

# Game loop
running = True
while running:
    clock.tick(10) 
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT and dx == 0:
                dx = -block_size
                dy = 0
            elif event.key == pygame.K_RIGHT and dx == 0:
                dx = block_size
                dy = 0
            elif event.key == pygame.K_UP and dy == 0:
                dy = -block_size
                dx = 0
            elif event.key == pygame.K_DOWN and dy == 0:
                dy = block_size 
                dx = 0
                
    snake_x += dx
    snake_y += dy

    if snake_x >= WIDTH or snake_x < 0 or snake_y >= HEIGHT or snake_y < 0:
        running = False

    screen.fill((0, 0, 0))


    pygame.draw.rect(screen, (255, 0, 0), (food_x, food_y, block_size, block_size))

    snake_head = [snake_x, snake_y]
    snake_body.append(snake_head)
    
    if len(snake_body) > snake_length:
        del snake_body[0]

    for block in snake_body[:-1]:
        if block == snake_head:
            running = False

    for block in snake_body:
        pygame.draw.rect(screen, (0, 255, 0), (block[0], block[1], block_size, block_size))

    if snake_x == food_x and snake_y == food_y:
        food_x, food_y = spawn_food()
        snake_length += 1

    pygame.display.update()

print("Game Over!")
print("Score", snake_length)
pygame.quit()
```
